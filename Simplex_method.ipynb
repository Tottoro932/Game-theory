## Симплекс метод

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import random 
```

```python
def matrix_gen (m,n,basis): # функция для генерации матрицы [m, n]
    A = np.zeros((m, n)) # итоговая матрица для работы метода
    col_zer = 0
    for i in range(m):
        for j in range(n):
            if i==0:
                A[i,j] = random.randint(-50, 10) 
            else:
                A[i,j] = random.randint(-50, 100) 
            
    A[0,n-1] = 0
    for i in range(len(basis)):
        stb = basis[i]-1
        for j in range(m):
            if j==i+1:
                A[j,stb] = 1
            else:
                A[j, stb] = 0
        
    return A
    
```

```python
def admis(A):  # проверка на то, является ли матрица двойственно допустимой
    k, l = A.shape
    fl = True
    E = 0.0001
    for i in range(l-1):
        if A[0,i] + E <0:
            fl = False
    return fl
```

```python
def find_c(A): # поиск ведущего столбца
    k, l = A.shape
    minim = 1
    nom = -1
    for i in range(l-1):
        if A[0,i]< minim:
            minim = A[0,i]
            nom = i
    return nom
```

```python
def find_l(A,j):  # поиск ведущей строки
    k, l = A.shape
    nom = -1
    for i in range(1,k):
        if  (A[i,j] != 0 ) and (A[i,l-1]/A[i,j])>0:
            if nom==-1:
                fl = A[i,l-1]/A[i,j]
                nom = i
            elif A[i,l-1]/A[i,j] < fl:
                fl = A[i,l-1]/A[i,j]
                nom = i
   
    return nom
```

```python
def change(A,i,j): # преобразовываем таблицу по Гауссу
    m, n = A.shape
    delit = 1/A[i,j]
    for p in range(n): # преобразовали строку
        A[i,p] = A[i,p] * delit 
    
    for t in range(m):
        mn = A[t,j] * (-1)
        if t!=i:
            for p in range(n):
                A[t,p] = A[t,p] + A[i,p] * mn
    return A
            
    
```

```python
def exm(A,basis): # проверка на существование решения
    m, n = A.shape
    fl_1 = True # решения существует
    for i in range(1,m):
        fl = True # для конкретной строки
        nom = basis[i-1]
        for j in range(n-1):
            if (j != nom-1) and (A[i,j]<0):
                fl = False
                
        if (fl==True) and (A[i,n-1]<0):
            fl_1 = False
    return fl_1
```

Постановка задачи:

```python
n = random.randint(3, 100) # кол-во столбцов матрицы (общее количество переменных + 1 стб)
m = random.randint(3, 100) # кол-во строк матрицы (кол-во базисных переменных + 1 стр)

if n<=m:
    if n == m:
        m = m-1
    else:
        n, m = m, n
print("Размерность будущей матрицы: ", m,n)

#variables = ['x'+str(i) for i in range(1,n)]
variables = [i for i in range(1,n)]
basis = np.random.choice(variables, m-1, replace = False)

print("Размер исходный базиса:", len(basis))
print("Исходный базис:", basis)

A = matrix_gen (m,n,basis) # сгенерировали матрицу 
print(A)
```

```python
fl =  admis(A)
while fl == False:
    if exm(A,basis)==False:
        print("Задача не имеет решения")
        break
    stb = find_c(A) # ведущий столбец
    print(stb)
    st = find_l(A,stb) # ведущая строка
    print(st)
    if st == -1:
        print("Функция неограниченно возрастает")
        break
    basis[st-1] = variables[stb] # поменяли базис
    A = change(A,st,stb) # поменяли матрицу
    print(A)
    print("Базис:",basis )
    print("----------------")
    fl =  admis(A)
```

```python
print("Значения: ")
for i in range(len(basis)):
    print("x", basis[i], "=", A[i+1,n-1])
print("Максимальное значение функции: ", A[0,n-1])
```
